```python
import numpy as np
import matplotlib.pyplot as plt

# Definicja funkcji LeakyReLU i jej gradientu
def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def leaky_relu_gradient(x, alpha=0.01):
    return np.where(x >= 0, 1, alpha)

# Generowanie danych
x = np.linspace(-5, 5, 500)
y = leaky_relu(x)
dy = leaky_relu_gradient(x)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))

# Wykres funkcji LeakyReLU
plt.plot(x, y, label='LeakyReLU (=0.01)', color='blue', linewidth=2)

# Wykres gradientu
plt.plot(x, dy, label='Gradient', color='red', linestyle='--', linewidth=2)

# Konfiguracja wykresu
plt.title('Funkcja Aktywacji LeakyReLU i jej Gradient', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.legend(fontsize=12)
plt.ylim(-0.05, 1.2)

# Wywietlenie wykresu
plt.show()
```
